```python
from sympy import *
from sympy.plotting import plot3d
from phi.api import *
```

```python
x, w, r = sp.symbols(['x', 'w', 'r'])
```

```python
p = 1 / ( 1 +  exp(-1 * w))

print simplify(diff(p, w)), "|", simplify(p * (1 - p)), "|", simplify(p * diff(log(p), w))
dpdw = 
g = dpdw * log(p) * r
plot(p, (w, -5, 5))
plot(dpdw, (w, -5, 5))
plot3d(g, (w, -5, 5), (r, -1, 1))
```
