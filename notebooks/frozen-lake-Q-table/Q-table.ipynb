```python
import numpy as np
import gym
from numpy.random import choice
import random

env = gym.make("FrozenLake-v0")
```

```python
def next_action(actions, e):
    n = len(actions)
    
    if random.random() < e:
        return random.randint(0, n-1)
    else:
        return np.argmax(actions)
```

```python
n_actions = env.action_space.n
n_states = env.observation_space.n
Q = np.zeros([n_states, n_actions])

learning_rate = 0.85
y = 0.95
k = 5000.0
```

```python
r_total = 0

for t in range(200000):
    s = env.reset()
    
    learning_rate = k / (k + t)
    
    while True:
        a = next_action(Q[s], 0.05)

        s_, r, done, info = env.step(a)
        r_total += r

        Q[s, a] += learning_rate * (r + y * np.max(Q[s_]) - Q[s, a])
        
        s = s_
        
        if done:
            break
            
    if t % 5000 == 0:
        print(r_total / 5000.0)
        r_total = 0
```

```python
s = env.reset()
done = False
for i in range(100):
    a = next_action(Q[s], 0)
    s, r, done, info = env.step(a)
    env.render()
    print("")
    
    if done:
        break
```
